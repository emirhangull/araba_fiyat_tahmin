```python
def fiyattemizle(stringim):
    stringim = stringim.replace(".","").replace("TL","").strip()
    if " " in stringim:
        stringim = stringim.split(" ")[-1]
    
    return stringim
```

```python
df.fiyat = df['fiyat'].apply(fiyattemizle)
```

```python
df.fiyat = df.fiyat.astype(int)
```

```python
# IQR hesapla
Q1 = df['fiyat'].quantile(0.25)
Q3 = df['fiyat'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Sınırlar dışındaki değerleri eşitle
df.loc[df['fiyat'] > upper_bound, 'fiyat'] = upper_bound
df.loc[df['fiyat'] < lower_bound, 'fiyat'] = lower_bound

```

```python
def mylabelencoder(stringim):
    if stringim == "orjinal":
        return 0
    elif stringim == "lokal_boyalı":
        return 20
    elif stringim == "boyalı":
        return 50
    elif stringim == "değişmiş":
        return 100
    elif stringim == "belirtilmemiş":
        return 40
    else:
        return np.nan

for col in df[araba_parcalari_listesi]:
    df[col] = df[col].apply(mylabelencoder)
```

```python
parca_korelasyonlari_fiyat = {}

for col in araba_parcalari_listesi:
    # Eğer kolon DataFrame'de yoksa atla (hatayı önlemek için)
    if col not in df.columns:
        print(f"Uyarı: '{col}' kolonu DataFrame'de bulunamadı ve atlandı.")
        continue
        
    # Korelasyonu hesapla
    correlation_value = df[col].corr(df['fiyat'])
    # Sözlüğe kaydet
    parca_korelasyonlari_fiyat[col] = correlation_value

# Sonuçları bir DataFrame'e dönüştürüp korelasyon değerine göre sıralayalım
correlation_results_df = pd.DataFrame.from_dict(parca_korelasyonlari_fiyat, orient='index', columns=['Fiyat Korelasyonu'])

# Korelasyonları mutlak değerine göre veya doğrudan artan/azalan değere göre sıralayabilirsiniz.
# Örneğin, en güçlü negatif korelasyonları (fiyatı en çok düşüren hasarları) görmek için:
correlation_results_df = correlation_results_df.sort_values(by='Fiyat Korelasyonu', ascending=True)

# Sonuçları ekrana yazdır
print("Her Bir Label Encoded Parça Kolonunun Fiyat ile Korelasyonu:")
print(correlation_results_df)
```

```python
# Her parça için korelasyonları bir sözlükte tut
korelasyon_dict = {
    "Sol Ön Çamurluk": -0.342187,
    "Motor Kaputu": -0.318984,
    "Sağ Ön Çamurluk": -0.304440,
    "Sol Arka Çamurluk": -0.291584,
    "Sol Arka Kapı": -0.289687,
    "Sağ Ön Kapı": -0.287189,
    "Sol Ön Kapı": -0.283878,
    "Sağ Arka Çamurluk": -0.250538,
    "Sağ Arka Kapı": -0.250244,
    "Arka Kaput": -0.217126,
    "Arka Tampon": -0.205353,
    "Ön Tampon": -0.204000,
    "Tavan": -0.170847
}

# Her satır için ağırlıklı hasar skoru hesapla
df["agirlikli_hasar_skoru"] = sum(df[parca] *(-1)* korelasyon_dict[parca] for parca in korelasyon_dict)
```

```python
corrbaba= df["agirlikli_hasar_skoru"].corr(df["fiyat"])
print("Ağırlıklı hasar skoru ile fiyat korelasyonu:", corrbaba)
```
